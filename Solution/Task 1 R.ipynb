# Read an excel file (DocumentData.xlsx)

```r
# Loading
library("readxl")
library("dplyr")
library("fuzzywuzzyR")
set.seed(123)
```

```r
# xlsx files
data= read_excel("../Data/DocumentsData.xlsx")
```

# Train-Test split

```r
train<-sample_frac(data, 0.7)
sid<-as.numeric(rownames(train))
test<-data[-sid,]
```

```r
dim(train)
```

```r
dim(test)
```

# Predict the best match of the test data docs from the training data docs

```r
my_list_train <- list()   
for(i in 1:nrow(train[,2])) { my_list_train[i] <- train[i,2]}
```

```r
my_list_test <- list()   
for(i in 1:nrow(test[,2])) { my_list_test[i] <- test[i,2]}
```

```r
choices=as.character(my_list_train)
```

```r
init_proc = FuzzUtils$new() 
PROC = init_proc$Full_process 
PROC1 = tolower      
init_scor = FuzzMatcher$new()    # initialization of the scorer class
SCOR = init_scor$WRATIO          # choosen scorer function
init <- FuzzExtract$new()        # Initialization of the FuzzExtract class
```

```r
word = as.character(my_list_test[1])
word
```

```r
# testing
sample=init$ExtractBests(string = word, sequence_strings = choices, processor = PROC, scorer = SCOR, score_cutoff = 95L, limit = 5L)
sample
```

```r
sample[[1]][2]
```

# Output matched doc, Count of matched docs and Confidence score

Fuzzywuzzy is used for Fuzzy text matching. Confidence score=100 indicates exact match. Based on Several execution it is observed that Score_cutoff = 95 shows same pattern in test and train data.

```r
best_matcher <- function(word) {
   init$ExtractBests(string = word, sequence_strings = choices, processor = PROC, scorer = SCOR, score_cutoff = 95L, limit = 1L)
}
best_matcher
```

```r
# Best match for test data from train data
mapply(best_matcher, test[["Doc Name"]])
```
